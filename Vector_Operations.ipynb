```python
# prompt: write sample code to plot addition of two vectors

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sms
sms.set()

# Define vectors
v1 = np.array([2, 3])
v2 = np.array([1, 2])

# Calculate the sum of vectors
v_sum = v1 + v2

# Plot vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1 + v2')

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Addition of Two Vectors')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Calculate the sum of vectors
v_sum = v1 - v2

# Plot vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1 - v2')

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Subtraction of Two Vectors')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the sum of vectors
v_sum = 2 * v2

# Plot vectors
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1 * v2')

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiplication of Two Vectors')
plt.legend()
plt.grid(True)
plt.show()
```
